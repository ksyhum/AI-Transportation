```python
import pandas as pd
```

```python
# import data, skiprows=1 karena data di csv mulai dari baris ke-2
dataset = pd.read_csv("Dataset-PT.csv",skiprows=1)
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
df = dataset[['arrival_delay', 'dwell_time', 'travel_time_for_previous_section',
       'scheduled_travel_time', 'upstream_stop_delay', 'origin_delay',
       'previous_bus_delay', 'previous_trip_travel_time', 'recurrent_delay', 'route_id', 'factor(weather)Light_Rain', 'factor(weather)Light_Snow', 'factor(weather)Normal', 'factor(weather)Rain', 
    'factor(weather)Snow', 'factor(temperature)Cold', 'factor(temperature)Extra_cold', 
    'factor(temperature)Normal', 'factor(day_of_week)weekday', 'factor(day_of_week)weekend', 
    'factor(time_of_day)Afternoon_peak', 'factor(time_of_day)Morning_peak', 
    'factor(time_of_day)Off-peak']]
```

```python
df.head()
```

```python
x = df.drop(['arrival_delay'], axis=1)
x.head()
```

```python
y = df['arrival_delay']
y.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

# Set default Seaborn style
sns.set()

# Membuat ukuran gambar lebih besar
plt.figure(figsize=(10, 6))

# Plot histogram dengan mengatur jumlah bin (interval sumbu x)
sns.histplot(x=df['arrival_delay'], bins=80) 

# Menampilkan grafik
plt.show()
```

```python
x.columns
```

```python
corr_matrix = df.corr()
corr_matrix['arrival_delay'].sort_values(ascending=False)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from xgboost import XGBRegressor
import xgboost as xgb
```

```python
#XGBoost Parameters Tuning

# not yet tuned
```

```python
model_XGB = XGBRegressor()
model_XGB.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
xgb.plot_importance(model_XGB, ax=plt.gca())
plt.show()
```

```python
# Predict the test data with the fitted model
y_pred_XGB = model_XGB.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae_XGB = mean_absolute_error(y_test, y_pred_XGB)
mse_XGB = mean_squared_error(y_test, y_pred_XGB)
r2_XGB = r2_score(y_test, y_pred_XGB)

print(f"Mean Absolute Error: {mae_XGB}")
print(f"Mean Squared Error: {mse_XGB}")
print(f"R-squared: {r2_XGB}")
```
